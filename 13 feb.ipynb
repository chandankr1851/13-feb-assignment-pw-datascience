```python
Q1. Explain why we have to use the Exception class while creating a Custom Exception.

In Python, the Exception class is the base class for all exceptions. When we create a custom exception,
it is recommended to subclass the built-in Exception class or one of its subclasses. By doing this, 
our custom exception inherits the behavior and functionality of the Exception class, 
making it consistent with other built-in exceptions in Python.

Using the Exception class as the base class ensures that our custom exception can be caught by generic 
except blocks that handle all types of exceptions. It also allows us to handle our custom exception
along with other standard exceptions in a consistent manner.

Example of creating a custom exception:
    class CustomError(Exception):
    pass

try:
    raise CustomError("This is a custom exception.")
except CustomError as e:
    print("Caught the custom exception:", e)
```

```python
Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

The ArithmeticError class is the base class for all numeric errors that occur during arithmetic operations in Python. 
Some errors defined in the ArithmeticError class are:

FloatingPointError: Raised when a floating-point arithmetic operation fails.
try:
    result = 1 / 0.0
except FloatingPointError as e:
    print("Caught FloatingPointError:", e)
OverflowError: Raised when the result of an arithmetic operation is too large to be expressed within the available numeric range.
import math

try:
    result = math.exp(1000)
except OverflowError as e:
    print("Caught OverflowError:", e)
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is the base class for all errors that occur when a key or index is not found in a collection,
such as dictionaries or lists. It provides a consistent way to handle situations where a lookup operation fails.

KeyError: Raised when a dictionary key is not found.
my_dict = {'a': 1, 'b': 2}

try:
    value = my_dict['c']
except KeyError as e:
    print("Caught KeyError:", e)
IndexError: Raised when an index is not found in a sequence (e.g., list, tuple, string).
my_list = [1, 2, 3]

try:
    value = my_list[3]
except IndexError as e:
    print("Caught IndexError:", e)
```
